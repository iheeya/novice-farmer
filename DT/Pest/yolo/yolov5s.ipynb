```python
import zipfile
import os

# 압축 파일 경로 설정
zip_file_path = '/content/drive/MyDrive/labels2.zip'

# 압축 해제 경로 설정 (해제된 파일을 저장할 위치)
extract_path = '/content/drive/MyDrive/custom_data'

# 디렉토리 생성 (이미 존재하는 경우 생략)
os.makedirs(extract_path, exist_ok=True)

# 압축 파일 해제
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# 압축 해제된 파일 경로 확인
print(f"Files extracted to: {extract_path}")
```
